```python
bins = [0, 100000, 200000, 300000, 400000, 500000, np.inf]
labels = ['<100k', '100k-200k', '200k-300k', '300k-400k', '400k-500k', '500k+']
df['PriceCategory'] = pd.cut(df['SalePrice'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='PriceCategory', order=df['PriceCategory'].value_counts().index)
plt.title('DistribuiÃ§Ã£o das categorias do Sale Price ')
plt.xlabel('Categoria dos preÃ§os')
plt.ylabel('Contagem')
plt.xticks(rotation=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop(['SalePrice', 'PriceCategory'], axis=1)
y = df['SalePrice']

X = pd.get_dummies(X, drop_first=True)

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

importances = rf.feature_importances_
feature_importances = pd.Series(importances, index=X.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
feature_importances.head(20).plot(kind='bar')
plt.title('Os 20 features mais importantes do Random Forest')
plt.xlabel('Features')
plt.ylabel('Importancia')
plt.show()
```
